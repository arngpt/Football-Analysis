```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

plt.scatter(df_players['overall_rating'],df_players['sliding_tackle'])
```

```python
df_test = df_players[['standing_tackle','finishing']]

df_test=df_test.dropna()
```

```python
kmeans = KMeans(4)
kmeans.fit(df_test)
```

```python
identified_clusters = kmeans.fit_predict(df_test)
identified_clusters
```

```python
data_with_clusters = df_test.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['standing_tackle'],data_with_clusters['finishing'],c=data_with_clusters['Clusters'],cmap='rainbow')
plt.xlabel('standing_tackle')
plt.ylabel('finishing')
```

```python
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(df_test)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
from sklearn.linear_model import LinearRegression

team_scoring['Team Name'] = team_scoring['Team Name'].astype(str)
```

```python
man_utd = team_scoring[team_scoring['Team Name']=='Manchester United ']
```

```python
import matplotlib.pyplot as plt

x = man_utd['Season']
y = man_utd['goals_scored'] - man_utd['goals_conceded']

plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder() 
x = label_encoder.fit_transform(x)
print(x)
```

```python
plt.title(x)
plt.scatter(x,y)
plt.plot(x, lin_reg.predict(x),color='k')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
fig = plt.figure(figsize=(10,15))
team_names = ['Manchester United ', 'Arsenal  ','Chelsea  ', 'Liverpool  ','Manchester City ','Tottenham Hotspur ']

for i in range(len(team_names)):
    team = team_scoring[team_scoring['Team Name']==team_names[i]]
    
    x = team['Season']
    y = team['goals_scored'] - team['goals_conceded']
    
    #to enconde season names
    label_encoder = LabelEncoder() 
    x = label_encoder.fit_transform(x)
    
    lin_reg = LinearRegression()
    x=x.reshape(-1, 1)
    lin_reg.fit(x,y)

    
    plt.subplot(2,3,i+1)  
    plt.scatter(team['Season'],y)
    plt.xticks(rotation=90)
    plt.title(team_names[i])
    plt.plot(team['Season'], lin_reg.predict(x),color='k')
```
